```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

```python
data = pd.read_excel('csv_companies/50_most_companies.xlsx')
```

```python
Ticker = data['Papel']
```

```python
#Ticker.to_csv('csv_companies/ticker.csv')
```

```python
# Function to fetch data for each ticker and return a DataFrame
def fetch_stock_data(Ticker):
    data = []
    for symbol in Ticker:
        stock = yf.Ticker(symbol+'.SA')
        df = stock.history(period="10d") # The amount of previews days
        df['Ticker'] = symbol
        data.append(df)
    return pd.concat(data, axis=0)

# Create the DataFrame
df = fetch_stock_data(Ticker)
df = df.reset_index()  # Reset the index to have a clean Date column

# Select and reorder the columns
df = df[['Ticker', 'Date', 'High', 'Low', 'Close', 'Volume']]

# Display the DataFrame
print(df)
```
