```python
import yfinance as yf
import streamlit as st
import pandas as pd
```

```python
data = pd.read_excel('csv_companies/ticker.csv')
```

```python
Ticker = data['Papel']
```

```python
# Function to fetch data for each ticker and return a DataFrame
def fetch_stock_data(Ticker):
    data = []
    for symbol in Ticker:
        stock = yf.Ticker(symbol+'.SA')
        df = stock.history(period="10d")
        df['Ticker'] = symbol
        data.append(df)
    return pd.concat(data, axis=0)

# Create the DataFrame
df = fetch_stock_data(Ticker)
df = df.reset_index()

# Select and reorder the columns
df = df[['Ticker', 'Date', 'High', 'Low', 'Close', 'Volume']]
df.to_csv('Company.csv')
# Display the DataFrame
print(df)

```

```python
# Group the DataFrame by 'Ticker'
grouped_df = df.groupby('Ticker')

# Group the DataFrame by 'Ticker'
grouped_df = df.groupby('Ticker')

# Create a dictionary to store DataFrames for each company
company_dataframes = {}

# Iterate through groups and create DataFrames
for ticker, group in grouped_df:
    company_dataframes[ticker] = group.copy()

# Save individual DataFrames to separate CSV files
for ticker, company_df in company_dataframes.items():
    file_name = f"{ticker}_stock_data.csv"
    company_df.to_csv(file_name, index=False)
    print(f"{ticker} DataFrame saved to {file_name}")
```
